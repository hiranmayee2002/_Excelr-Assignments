```python
import pandas as pd

# Load the Excel file
data = pd.read_excel(r'C:\Users\Dell\Desktop\Data Science Assignments\Association Rules\Online retail.xlsx')

# Ensure each transaction (row) has a unique index
data = data.reset_index(drop=True)

```

```python
# Split the items into lists and remove any duplicates within each transaction
data['items'] = data.iloc[:, 0].apply(lambda x: list(set(x.split(','))))

```

```python
data.head()
```

```python
#the list of all items to get unique items
all_items = sorted(set(item for sublist in data['items'] for item in sublist))
```

```python
# Initialize an empty DataFrame with columns as unique items
basket = pd.DataFrame(0, index=data.index, columns=all_items)

```

```python
# Populate the basket with 1s where applicable
for i, items in enumerate(data['items']):
    basket.loc[i, items] = 1
print(basket.head())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Applyying the Apriori algorithm
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

# Generating the  association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sorting of  rules by confidence and lift
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])

rules
```

```python
# Check for duplicate columns
duplicate_columns = basket.columns[basket.columns.duplicated()]

if len(duplicate_columns) > 0:
    print(f"Found duplicate columns: {duplicate_columns}")
    # Handle duplicates if necessary, e.g., by dropping them
    basket = basket.loc[:, ~basket.columns.duplicated()]
duplicate_columns
```
